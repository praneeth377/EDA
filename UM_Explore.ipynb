```python
df2['Event name'] = df2['Event name'].str.split('(').str.get(0)
```

#### Cleaning up Athlete Age

```python
df2['Athlete age category'] = df2['Athlete age category'].str[-2:]
```

```python
df2['Athlete age category'] = df2['Athlete age category'].fillna(0).astype('int')
```

```python
df2.rename(columns={'Athlete age category':'Athlete Age'}, inplace = True)
```

```python
df2.head()
```

#### Removing 'h' from Athlete Performance

```python
df2['Athlete performance'] = df2['Athlete performance'].str.split(' h').str.get(0)
```

```python
df2.head()
```

#### Dropping unwanted columns

```python
df2.drop(columns=['Athlete club','Athlete country','Athlete year of birth'], axis=1, inplace=True)
```

```python
df2.head()
```

#### Checking for Null values

```python
df2.isna().sum()
```

#### Checking for Duplicate entries

```python
df2[df2.duplicated()]
```

```python
df2.reset_index(drop=True, inplace=True)
```

#### Fixing Data Types for each column

```python
df2.dtypes
```

```python
df2['Athlete average speed'] = df2['Athlete average speed'].astype('float64')
```

#### Cleaning the Date column and converting the data type to 'datetime'

```python
df2[df2['Event dates'] == '18.-19.01.2020'].head(3)
```

```python
df2['Event dates'] = df2['Event dates'].apply(lambda x : x[5:] if (len(x) > 10) else x)
```

```python
df2['Event dates'] = pd.to_datetime(df2['Event dates'], format='%d.%m.%Y', exact=False)
```

### Questions I want to answer from Data

#### 1. Difference in speed for the 50k, 50mi male to female

```python
df3.groupby(['Race_Length','Athlete gender'])['Athlete average speed'].agg('mean').to_frame()
```

##### No drastic difference between genders.

#### 2. What Age groups are best in 50mi Race

```python
df3.query('Race_Length == "50mi"').groupby('Athlete Age')['Athlete average speed'].agg(['mean','count']).sort_values('mean', ascending=False)
```

##### Ignoring the missing values which have been imputed with 0 earlier, we can see ascending order of Athlete Ages in index column. This again corroborates to earlier 'lmplot' that as age increases, mean speeds drop. 

#### 3. Seasons for the data : 

#### Spring 3-5
#### Summer 6-8
#### Fall 9-11
#### Winter 12-2

#### Are athletes slower in summer than winter ?

```python
# Creating a new column 'Race_Season'
df3['Race_Season'] = ''
```

```python
conditions = [
    (df3['Race_Day'].dt.month >= 3) & (df3['Race_Day'].dt.month <= 5),
    (df3['Race_Day'].dt.month >= 6) & (df3['Race_Day'].dt.month <= 8),
    (df3['Race_Day'].dt.month >= 9) & (df3['Race_Day'].dt.month <= 11),
]

seasons = ['Spring', 'Summer', 'Fall']

# Using numpy.select() to assign the season based on conditions
df3['Race_Season'] = np.select(conditions, seasons, default='Winter')
```

```python
df3.head()
```

```python
df3.groupby('Race_Season')['Athlete average speed'].agg(['mean','count']).sort_values('mean', ascending=False)
```

##### We observe that, on average, athletes tend to be slower in the summer compared to winter. However, it's important to note that this observation is based on the available data, which comprises over 10,000 records for the winter season and only around 3000 for summer. While this pattern holds in our dataset, it's crucial to recognize that the sample size for summer is significantly smaller, and generalizing this trend to the broader population should be done cautiously, as real-world variations may exist.
